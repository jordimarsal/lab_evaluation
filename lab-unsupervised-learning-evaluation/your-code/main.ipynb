Load the wine dataset using sklearn and get the datasets X and y containing the target and the rest of the variables

```python
from sklearn.datasets import load_wine

wine = load_wine()
X = wine["data"]
Y = wine["target"]
names = wine["feature_names"]
```

```python
X = pd.DataFrame(X, columns=names)
X.tail()
```

## 1. Train a KMeans clustering model on the data set using 8 clusters and compute the silhouette score for the model.

```python
from sklearn.metrics import silhouette_score
from sklearn.cluster import KMeans

kmeans8 = KMeans(n_clusters=8, random_state=42)
wine_8 = kmeans8.fit_predict(X)
```

```python
s8 = silhouette_score(X,wine_8)
print(s8)
```

## 2. Train a KMeans clustering model on the data set using 5 clusters and compute the silhouette score for the model.

```python
kmeans5 = KMeans(n_clusters=5, random_state=42)
wine_5 = kmeans5.fit_predict(X)
s5 = silhouette_score(X,wine_5)
print(s5)
```

## 3. Train a KMeans clustering model on the data set using 3 clusters and compute the silhouette score for the model.

```python
kmeans3 = KMeans(n_clusters=3, random_state=42)
wine_3 = kmeans5.fit_predict(X)
s3 = silhouette_score(X,wine_3)
print(s3)
```

## 4. Use elbow curve visualizations to see if you can determine the best number of clusters to use.

The Yellowbrick library has 3 metrics that you can plot using the `metric` parameter:

- **distortion**: mean sum of squared distances to centers
- **silhouette**: mean ratio of intra-cluster and nearest-cluster distance
- **calinski_harabaz**: ratio of within to between cluster dispersion

```python
#!pip install yellowbrick
from yellowbrick.cluster import KElbowVisualizer
```

```python
def elbow(model, X, ):
    visualizer = KElbowVisualizer(model, k=(4,12),metric='distortion')

    visualizer.fit(X)        # Fit the data to the visualizer
    visualizer.show()  
```

```python
elbow(kmeans8, X)
elbow(kmeans5, X)
elbow(kmeans3, X)
```

## 7. Apply a PCA transform and plot the first two principle components with the plot point colors determined by cluster.

```python
from sklearn.decomposition import PCA

pca = PCA(n_components=2)
principalComponents = pca.fit_transform(X)

principalDf = pd.DataFrame(data = principalComponents
             , columns = ['principal component 1', 'principal component 2'])
```

```python
with sns.color_palette("husl", 8):
    sns.scatterplot(x='principal component 1', y='principal component 2',
                hue=X['labels'],data=principalDf)
```
